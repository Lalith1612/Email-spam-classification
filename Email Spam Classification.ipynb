# Import Libraries

```python
word_columns = ['word_freq_make', 'word_freq_address', 'word_freq_all', 'word_freq_3d', 'word_freq_our']

plt.figure(figsize=(12, 8))

# Plotting histograms for selected word frequencies
for i, col in enumerate(word_columns):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

This will help us see how common or rare certain words are in the dataset. The shape of the histogram (e.g., skewness) might help us determine if most emails have high or low occurrences of specific words, like "make", "address", "3d", etc.

```python
# Compute correlation matrix for word frequency columns
correlation_matrix = data[word_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Word Frequencies')
plt.show()
```

Strong positive correlations (values close to 1) between certain words indicate that these words might appear in the same types of emails (for example, "money" and "credit" may correlate strongly).

Negative correlations might indicate competing features or the presence of disjoint content (e.g., "free" and "business").

```python
# Example: Boxplot to show word frequencies distribution for each class (spam: 1, non-spam: 0)
plt.figure(figsize=(12, 8))

# Plotting a boxplot for selected word frequency features
for i, col in enumerate(word_columns):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x=data['class'], y=data[col])
    plt.title(f'{col} vs. Class')

plt.tight_layout()
plt.show()
```

Boxplots will show the spread of word frequencies for both classes (class = 1 for spam, class = 0 for non-spam).
                                                 
You may see certain words such as "money", "free", and "credit" having higher frequencies in spam emails.

```python
# Histogram for capital run lengths
capital_columns = ['capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total']

plt.figure(figsize=(12, 8))

# Plotting histograms for capital run length features
for i, col in enumerate(capital_columns):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[col], kde=True, bins=20, color='green')
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

Capitalized words may appear more frequently in spam emails, often used to attract attention. Analyzing how many times there is a long stretch of capitalized letters or a high number of capitalized words could be useful in distinguishing spam messages from non-spam ones.

```python
# Selected features to compare
selected_features = ['word_freq_free', 'word_freq_money', 'word_freq_you', 'capital_run_length_average']

plt.figure(figsize=(12, 10))

for i, col in enumerate(selected_features):
    plt.subplot(2, 2, i + 1)
    sns.kdeplot(data=data, x=col, hue='class', fill=True, palette='Set2', alpha=0.6)
    plt.title(f'Density Plot: {col} by Class')
    plt.xlabel(col)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

The overlap or separation in density curves between spam (class=1) and non-spam (class=0) will indicate how well the feature can distinguish between these two classes.

```python
# Reduced feature set for visualization due to high dimensionality
features_to_plot = ['word_freq_free', 'word_freq_you', 'capital_run_length_average', 'capital_run_length_total', 'class']
sns.pairplot(data[features_to_plot], hue='class', diag_kind='kde', palette='husl', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Selected Features Grouped by Class', y=1.02)
plt.show()
```

Pairplots reveal correlations or clustering patterns among features.

Different clusters may appear for spam and non-spam emails, helping you see separability visually.

```python
# Calculate the class distribution
class_distribution = data['class'].value_counts()

# Labels for the pie chart
labels = ['Non-Spam', 'Spam']

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(class_distribution, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'], explode=(0, 0.1))
plt.title('Proportion of Spam vs. Non-Spam Emails')
plt.show()
```

This chart provides a quick overview of the class balance in your dataset.

If the dataset is highly imbalanced (e.g., one class dominates), it may influence model training and require special handling like resampling techniques or stratified splits.

```python
# Calculate the mean of selected word frequencies for each class
selected_features = ['word_freq_make', 'word_freq_address', 'word_freq_mail', 'word_freq_email', 'word_freq_free']
avg_word_freq_by_class = data.groupby('class')[selected_features].mean().T

# Plot a line chart
plt.figure(figsize=(10, 6))
for cls in avg_word_freq_by_class.columns:
    plt.plot(avg_word_freq_by_class.index, avg_word_freq_by_class[cls], marker='o', label='Spam' if cls == 1 else 'Non-Spam')

# Add chart details
plt.title('Average Word Frequency by Class', fontsize=14)
plt.xlabel('Word Features', fontsize=12)
plt.ylabel('Average Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Class', loc='upper right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

The chart shows how frequently specific words occur in spam vs. non-spam emails.
    
For example, if "free" or "email" has significantly higher averages for spam emails, these words can be strong indicators for spam classification.

# Show input X and output y

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

# Show input X

```python
X
```

# Show output y

```python
y
```

# MinMaxScaler for Data

```python
scaler = MinMaxScaler(copy=True, feature_range=(0, 1))
X = scaler.fit_transform(X)
print('X \n' , X[:10])
print('y \n' , y[:10])
```

# Splitting data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# Applying LogisticRegression Model 

```python
LogisticRegressionModel = LogisticRegression(penalty='l2',solver='sag',C=1.0,random_state=33)
LogisticRegressionModel.fit(X_train, y_train)
```

# Calculating Details

```python
print('LogisticRegressionModel Train Score is : ' , LogisticRegressionModel.score(X_train, y_train))
print('LogisticRegressionModel Test Score is : ' , LogisticRegressionModel.score(X_test, y_test))
print('LogisticRegressionModel Classes are : ' , LogisticRegressionModel.classes_)
print('LogisticRegressionModel No. of iteratios is : ' , LogisticRegressionModel.n_iter_)
```

```python
MLPClassifierModel = MLPClassifier(activation='relu', # can be also identity , logistic , relu
                                   solver='lbfgs',  # can be also sgd , adam
                                   learning_rate='constant', # can be also invscaling , adaptive
                                   early_stopping= False,
                                   alpha=0.0001 ,hidden_layer_sizes=(100, 10),random_state=33)
MLPClassifierModel.fit(X_train, y_train)
```

# Calculating Details

```python
print('MLPClassifierModel Train Score is : ' , MLPClassifierModel.score(X_train, y_train))
print('MLPClassifierModel Test Score is : ' , MLPClassifierModel.score(X_test, y_test))
print('MLPClassifierModel loss is : ' , MLPClassifierModel.loss_)
print('MLPClassifierModel No. of iterations is : ' , MLPClassifierModel.n_iter_)
print('MLPClassifierModel No. of layers is : ' , MLPClassifierModel.n_layers_)
print('MLPClassifierModel last activation is : ' , MLPClassifierModel.out_activation_)
```

# Applying RandomForestClassifier Model 

```python
RandomForestClassifierModel = RandomForestClassifier(criterion = 'gini',n_estimators=100,max_depth=20,random_state=33) #criterion can be also : entropy 
RandomForestClassifierModel.fit(X_train, y_train)
```

# Calculating Details

```python
print('RandomForestClassifierModel Train Score is : ' , RandomForestClassifierModel.score(X_train, y_train))
print('RandomForestClassifierModel Test Score is : ' , RandomForestClassifierModel.score(X_test, y_test))
print('RandomForestClassifierModel features importances are : ' , RandomForestClassifierModel.feature_importances_)
```

# Applying DecisionTreeClassifier Model 

```python
DecisionTreeClassifierModel = DecisionTreeClassifier(criterion='gini',max_depth=20,random_state=33) #criterion can be entropy
DecisionTreeClassifierModel.fit(X_train, y_train)
```

# Calculating Details

```python
print('DecisionTreeClassifierModel Train Score is : ' , DecisionTreeClassifierModel.score(X_train, y_train))
print('DecisionTreeClassifierModel Test Score is : ' , DecisionTreeClassifierModel.score(X_test, y_test))
print('DecisionTreeClassifierModel Classes are : ' , DecisionTreeClassifierModel.classes_)
print('DecisionTreeClassifierModel feature importances are : ' , DecisionTreeClassifierModel.feature_importances_)
```

# Applying KNeighborsClassifier Model 

```python
KNNClassifierModel = KNeighborsClassifier(n_neighbors=15,weights ='distance', # it can be distance
                                          algorithm='auto') # it can be ball_tree, kd_tree,brute
KNNClassifierModel.fit(X_train, y_train)
```

# Calculating Details

```python
print('KNNClassifierModel Train Score is : ' , KNNClassifierModel.score(X_train, y_train))
print('KNNClassifierModel Test Score is : ' , KNNClassifierModel.score(X_test, y_test))
```

# Display DecisionTreeClassifierModel.feature_importances_

```python
x_bar=list(range(1,58))
plt.bar(x_bar,DecisionTreeClassifierModel.feature_importances_)
```

# Calculating Prediction

```python
y_pred = RandomForestClassifierModel.predict(X_test)
y_pred_prob = RandomForestClassifierModel.predict_proba(X_test)
print('Predicted Value for RandomForestClassifierModel is : ' , y_pred[:10])
print('Prediction Probabilities Value for RandomForestClassifierModel is : ' , y_pred_prob[:10])
```

# Calculating Confusion Matrix

```python
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)
sns.heatmap(CM, center = True)
plt.show()
```

# Calculating Confusion Matrix

```python
CM = confusion_matrix(y_train,RandomForestClassifierModel.predict(X_train))
print('Confusion Matrix is : \n', CM)
sns.heatmap(CM, center = True)
plt.show()
```

# Calculating Accuracy Score : ((TP + TN) / float(TP + TN + FP + FN))

```python
AccScore = accuracy_score(y_test, y_pred, normalize=True)
print('Accuracy Score is : ', AccScore)
```

# Calculating F1 Score : 2 (precision recall) / (precision + recall)

```python
F1Score = f1_score(y_test, y_pred, average='micro')
print('F1 Score is : ', F1Score)
```

# Calculating Recall Score : (Sensitivity) (TP / float(TP + FN)) 1 / 1+2

```python
RecallScore = recall_score(y_test, y_pred, average='micro') #it can be : binary,macro,weighted,samples
print('Recall Score is : ', RecallScore)
```

# Calculating Precision Score : (Specificity) #(TP / float(TP + FP))

```python
PrecisionScore = precision_score(y_test, y_pred, average='micro')
print('Precision Score is : ', PrecisionScore)
```

# Calculating Precision recall Curve 

```python
PrecisionValue, RecallValue, ThresholdsValue = precision_recall_curve(y_test,y_pred)
print('Precision Value is : ', PrecisionValue)
print('Recall Value is : ', RecallValue)
print('Thresholds Value is : ', ThresholdsValue)
```

# Calculating classification Report

```python
ClassificationReport = classification_report(y_test,y_pred)
print('Classification Report is : ', ClassificationReport )
```

# Calculating ROC AUC Score:

```python
ROCAUCScore = roc_auc_score(y_test,y_pred, average='micro') #it can be macro,weighted,samples
print('ROCAUC Score : ', ROCAUCScore)
```
